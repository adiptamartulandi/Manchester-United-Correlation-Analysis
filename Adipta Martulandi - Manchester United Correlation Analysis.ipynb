## 5. Data Preprocessing

### 5.1 Handling Missing Values

```python
utd.isnull().sum()
```

```python
utd['saves'] = utd['saves'].fillna(0)
```

```python
utd['head_clearance'] = utd['head_clearance'].fillna(686)
```

```python
utd['total_through_ball'] = utd['total_through_ball'].fillna(utd['total_through_ball'].median())
```

```python
utd['dispossessed'] = utd['dispossessed'].fillna(utd['dispossessed'].median())
```

### 5. Remove Unused Columns

```python
dropped = ['backward_pass', 'big_chance_missed', 'team', 'total_yel_card', 'total_red_card', 'hit_woodwork',
           'total_offside', 'saves', 'own_goals', 'corner_taken', 'big_chance_missed', 'penalty_save', 'penalty_save'
          ,'total_high_claim', 'punches','att_hd_goal', 'att_pen_goal', 'att_freekick_goal', 'att_ibox_goal',
           'att_obox_goal','penalty_conceded', 'pen_goals_conceded', 'head_clearance', 'clearance_off_line']
utd.drop(columns=dropped, inplace=True)
```

```python
utd.head()
```

## 6. Feature Extraction

```python
#Membuat fitur block, interception dan clearence
utd['defence'] = utd['outfielder_block'] + utd['interception'] + utd['total_clearance']
utd.drop(columns=['outfielder_block', 'interception', 'total_clearance'], inplace=True)
```

```python
#Membuat fitur draw
#1 Musim ada 38 pertandingan
utd['draws'] = 38 - utd['wins'] - utd['losses']
```

```python
#Rearange columns agar mudah untuk dibaca
utd = utd[['wins', 'losses', 'draws', 'goals', 'goal_fastbreak', 'total_scoring_att', 'ontarget_scoring_att',
          'total_pass', 'total_through_ball', 'total_long_balls', 'total_cross', 'touches', 'clean_sheet', 'dispossessed',
          'goals_conceded', 'total_tackle', 'last_man_tackle', 'defence', 'season']]
```

### 7.1 General Analysis
- Features yang digunakan adalah:
    1. wins = Jumlah MU menang
    2. losses = Jumlah MU kalah

```python
plt.figure(figsize=(10,5))

plt.plot(utd['season'], utd['wins'])
plt.plot(utd['season'], utd['losses'])
plt.xlabel('Season')
plt.ylabel('Numbers')
plt.xticks(rotation=45)

plt.legend(labels=['wins', 'losses'])
plt.show()
```

- Secara <b>general</b>, dapat dilihat bahwa sebelum sir alex era MU memiliki kemenangan diatas sekitar 23 pertandingan pada 1 musim.
- Lalu setelah sir alex era kemenangan MU mulai turun bahkan sampai di bawah angka 20 pertandingan pada 1 musim.
- Kecuali tahun 2017/2018 yang memiliki kemenangan diatas 20.

### 7.2 Attacking Analysis
- Menggunakan 4 features yaitu:
    1. goals = total goals
    2. goal_fastbreak = goals dalam counter attack
    3. total_scoring_att = total shots
    4. ontarget_scoring_att = total shots on target

```python
#Analisis goalsnya dulu
plt.figure(figsize=(10,5))

plt.plot(utd['season'], utd['goals'])
plt.plot(utd['season'], utd['goal_fastbreak'])
plt.xlabel('Season')
plt.ylabel('Numbers')
plt.xticks(rotation=45)

plt.legend(labels=['total goals', 'counter attack'])
plt.show()
```

- Pada era sir alex total goalsnya di kisaranangka 70-80 goals, lalu setelah sir alex pensiun total goalsnya  langsung turun di kisaran angka 50-60 pada satu musim.
- Untuk goals dalam counter attack lebih banyak terjadi di era sir alex.
- Namun terjadi penurunan yang smooth untuk counter attack goals mulai 2008/2009, mungkin karena Cristiano Ronaldo pindah ?

```python
#Analisis Shoot
plt.figure(figsize=(10,5))

plt.plot(utd['season'], utd['total_scoring_att'], color='green')
plt.plot(utd['season'], utd['ontarget_scoring_att'], color='red')
plt.xlabel('Season')
plt.ylabel('Numbers')
plt.xticks(rotation=45)

plt.legend(labels=['total shoot', 'shoot on target'])
plt.show()
```

- Cukup jelas ya kalo pada era sir alex total shoots kisaran di angka 600-700 lalu setelah sir alex mengalami penurunan di angka 500an.
- Kecuali pada tahun 2016-2017 yang memiliki total shoot sekitar 600.

```python
#lihat korelasi
utd[['wins', 'goals', 'goal_fastbreak','total_scoring_att',
     'ontarget_scoring_att']].corr().style.background_gradient().set_precision(2)
```

- Dengan mengetahui korelasi diatas dapat diketahui bahwa ternyata hubungan antara kemenangan dan jumlah shot itu ternyata positif dan cukup kuat.
- Dengan kata lain dengan bertambahnya jumlah shot maka kemungkinan besar MU akan menang akan bertambah.
***
- Mungkin dari sini bisa dilihat bahwa MU harus mencari pemain penyerang ataupun gelandang yang berani shot atau memiliki akurasi shot yang tinggi.
- Di era sir alex, banyak sekali pemain dengan tipikal shooter seperti scholes, CR7, Tevez, Carrick, Rooney.

### 7.3 Midfielder Analysis
- Analisis akan menggunakan 3 features:
    1. total_pass = jumlah passes
    2. total_through_ball = jumlah umpan terobosan
    3. total_cross = jumlah crossing

```python
#Analisis total pass, through ball, dan long balls
f = plt.figure(figsize=(20,5))

f.add_subplot(1,2,1)
sns.lineplot(utd['season'], utd['total_pass'])
plt.xticks(rotation=45)

f.add_subplot(1,2,2)
sns.lineplot(utd['season'], utd['total_through_ball'])
plt.xticks(rotation=45)

f.tight_layout()
plt.show()
```

- Wow Ternyata semakin tahun nilai total_passes semakin naik, tetapi kok total_goalsnya semakin dikit ya ? kalau menurut saya ada 2 kemungkinan yaitu:
    1. penyelesaian akhir yang kurang bagus dari pemain depan MU atau mungkin memang MU menginisiasi untuk bermain possession ball.
- Lalu untuk umpan terobosan dapat dilihat bahwa mempunyai trend yang menurun. Walaupun total pass MU semakin naik, tapi umpan terobosan malah semakin turun. Menurut saya ada 2 kemungkinan juga yaitu:
    1. Semakin modern sepakbola semakin rapi pertahanan dan susah untuk melakukan through pass.
    2. Midfielder dari MU kurang kreativitas untuk mencari celah pertahanan.

```python
#Analisis total_cross
plt.figure(figsize=(10,5))

plt.plot(utd['season'], utd['total_cross'])
plt.xlabel('Season')
plt.ylabel('Numbers')
plt.xticks(rotation=45)

plt.legend(labels=['total cross'])
plt.show()
```

- Dapat dilihat bahwa setelah sir alex pensiun, jumlah crossing menurun sangat signifikan yak. Disini ada 2 kemungkinan menurut analisis saya:
    1. Gaya permainan pelatih setelah sir alex beda, mungkin david moyes atau van gall lebih suka bermain dari tengah sedangkan sir alex lebih suka memanfaatkan attacking dari winger.
    2. Setelah ditinggal CR7, Giggs, Nani mungkin belum ada pengganti winger yang cocok.

```python
#lihat korelasi
utd[['wins', 'total_pass', 'total_through_ball', 'total_cross']].corr().style.background_gradient().set_precision(2)
```

- Dari nilai korelasi diatas mungkin bisa ditarik beberapa kesimpulan yaitu:
    1. Ternyata total pass mempunyai korelasi negative dengan wins. Ini mungkin karena passing di MU kurang efektif hanya passing-passing yang tidak jelas dan akhirnya tidak menjadi goal.
    2. Untuk umpan terobosan, mempunyai korelasi yang positif dan cukup kuat. Sehingga perlu menambah umpan terobosan agar bisa mencetak lebih banyak goals dan bisa menang.
    3. Yap jumlah wins dan total_cross mempunyai korelasi positif yang sangat kuat dimana semakin sering melakukan crossing bola ke depan gawang maka akan semakin besar kemungkinan untuk mencetak goal dan meningkatkan peluang MU untuk menang.
***
- Jadi menurut analisis saya, MU harus mencari gelandang yang kreatif dan memiliki visi yang bagus. Seperti Hazard, Kevin de Bruyne atau Luka Modric.
- Tidak hanya bagus di passing, tetapi juga pintar mencari celah untuk memberikan umpan terobosan.
- Menambah kekuatan di sisi winger menurut saya bisa menambah peluang untuk meraih kemenangan.
